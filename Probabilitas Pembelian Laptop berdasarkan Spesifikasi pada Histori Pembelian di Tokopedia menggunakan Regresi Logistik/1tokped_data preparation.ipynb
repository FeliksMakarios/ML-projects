### Mengurai data

```python
#Menambahkan kolom kategori laptop berdasarkan kolom name yang mengandung sebagian string
cond_list = [laptop['name'].str.contains('acer', case=False),
             laptop['name'].str.contains('asus', case=False),
             laptop['name'].str.contains('dell', case=False),
             laptop['name'].str.contains('hp', case=False),
             laptop['name'].str.contains('lenovo', case=False),
             laptop['name'].str.contains('msi', case=False)]
choice_list = ['Acer', 'Asus', 'Dell', 'HP', 'Lenovo', 'MSI']

laptop['Category'] = np.select(cond_list, choice_list)
laptop.head()
```

```python
sold_mean = laptop.groupby(['location', 'date']).agg({'sold': np.sum})
sold_mean = sold_mean.reset_index()
print(sold_mean)
```

```python
grouped = sold_mean.groupby('date')

ncols=2
nrows = int(np.ceil(grouped.ngroups/ncols))

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12,32), sharey=True)
#plt.tight_layout()
fig.subplots_adjust(hspace=2)

for (key, ax) in zip(grouped.groups.keys(), axes.flatten()):
    grouped.get_group(key).plot(x='location', y='sold', ax=ax)
    ax.set_title("Tanggal {}".format(key))
    #ax.xaxis.set_tick_params(labelbottom=True)
    #ax.yaxis.set_tick_params(labelleft=True)
    #ax.set_xticks(np.arange(len(x)))
    
ax.legend(loc = 'best')
#plt.savefig('./dataset/baru/test.png')
plt.show()
```

```python
# menampilkan grafik total jumlah penumpang dari setiap kereta pada setiap stasiun
grouped = sold_mean.groupby(['date'])
fig, ax = plt.subplots(figsize=(16,8))
for key, group in grouped:
    group['sold'].plot(label=key, ax=ax)

plt.legend(loc='best')
#plt.savefig('./dataset/baru/test2.png')
plt.show()
```
